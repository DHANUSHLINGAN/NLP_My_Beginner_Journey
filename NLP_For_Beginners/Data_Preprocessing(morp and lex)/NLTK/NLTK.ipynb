POS Tagging and Stop Words

```python
nltk.download("averaged_perceptron_tagger")
```

```python
for token in ml_tokens:
  print(nltk.pos_tag([token]))
```

```python
sent = "jerry eats a banana"
tokens = nltk.word_tokenize(sent)
for token in tokens:
    print(nltk.pos_tag([token]))
```

```python
from nltk.tokenize import RegexpTokenizer
reg_tokenizer = RegexpTokenizer('(?u)\W+|\$[\d\.]+|\S+')
tokens = reg_tokenizer.tokenize(sent)
for token in tokens:
    print(nltk.pos_tag([token]))
```

Stemming

```python
#porter Stemmer
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
words = ['generous','generate','genorously','generation']
for word in words:
  print(f"{word} --> {porter.stem(word)}")
```

```python
#Snowball Stemmer
from nltk.stem import SnowballStemmer
porter = SnowballStemmer(language = 'english')
words = ['generous','generate','genorously','generation']
for word in words:
  print(f"{word} --> {porter.stem(word)}")
```

```python
#LancasterStemmer
from nltk.stem import LancasterStemmer
porter = LancasterStemmer()
words = ['generous','generate','genorously','generation']
for word in words:
  print(f"{word} --> {porter.stem(word)}")
```

```python
#LancasterStemmer
from nltk.stem import RegexpStemmer
porter = RegexpStemmer('ing|s$|ables$',min=4)
words = ['generous','generate','genorously','generation']
for word in words:
  print(f"{word} --> {porter.stem(word)}")
```

Lemmetization

#Word Sense Disambiguation

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
a1 = lesk(word_tokenize('The building has a device to jam the signal'),'jam')
print(a1,a1.definition())
a2 = lesk(word_tokenize("I'am stuck in a traffic"),"jam")
print(a2,a2.definition())
```

```python
a1 = lesk(word_tokenize("What is the current time?"),'current')
print(a1,a1.definition())
a2 = lesk(word_tokenize('air_current'),'current')
print(a2,a2.definition())
```

Named Entity Recognition

```python
nltk.download('maxenet_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
#Example 1
text = "Apple is an American company based on of California"
for w in nltk.word_tokenize(text):
    for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(w))):
        if hasattr(chunk,'label'):
            print(chunk.label(),''.join(c[0] for c in chunk))#Prints the Geo Political Entity(GPE)
```

```python
#Example 2
text = "Narendra Modi is the Prime minister of India"
for w in nltk.word_tokenize(text):
    for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(w))):
        if hasattr(chunk,'label'):
            print(chunk.label(),''.join(c[0] for c in chunk))#Prints the Geo Political Entity(GPE)


##You can also Train the model and after that you can post the pos tagging for better version which has less acquracy
```
