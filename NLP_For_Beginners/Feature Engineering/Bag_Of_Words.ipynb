```python
import pandas as pd
import numpy as np
import collections
```

```python
doc1 = "Harry Potter is an amazing movie!!"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry Potter is so great"
```

```python
import re
doc1 = re.sub(r"[^a-zA-Z0-9]"," ",doc1.lower()).split()
doc2 = re.sub(r"[^a-zA-Z0-9]"," ",doc2.lower()).split()
doc3 = re.sub(r"[^a-zA-Z0-9]"," ",doc3.lower()).split()
```

```python
doc1
```

```python
doc2
```

```python
allwords = set(doc1+doc2+doc3)
```

```python
def BOWrepresentation(all_words,doc):
    bow  = dict.fromkeys(all_words,0)
    for word in doc:
        bow[word] = doc.count(word)
    return bow
```

```python
bow1 = BOWrepresentation(allwords,doc1)
bow1
```

```python
bow2 = BOWrepresentation(allwords,doc2)
bow2
```

```python
bow3 = BOWrepresentation(allwords,doc3)
bow3
```

```python
df_bow = pd.DataFrame([bow1,bow2,bow3])
df_bow
```

#Using Scikit Learn


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary = True)#Sentiment Classification
doc1 = "Harry Potter is an amazing movie!!"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry Potter is so great"
cv_out = cv.fit_transform([doc1,doc2,doc3])
```

```python
cvout = cv_out.toarray()
```

```python
pd.DataFrame(cvout,columns = cv.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range = (1,3)) # Counting of how many times a word is being used = unigram,bigram,trigram
doc1 = "Harry Potter is an amazing movie!!"
doc2 = "Harry Potter is the best movie!"
doc3 = "Harry Potter is so great"
cv_out = cv.fit_transform([doc1,doc2,doc3])
```

```python
pd.DataFrame(cv_out.toarray(),columns = cv.get_feature_names_out())
```
